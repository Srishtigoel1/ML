```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn import linear_model
```

```python
df=pd.read_csv("Price.csv")
df
```

```python
%matplotlib inline
plt.scatter(df.Area,df.Price,color='red',marker='+')
plt.xlabel('area')
plt.ylabel('price')
```

```python
reg=linear_model.LinearRegression()
#for training linear regression model
reg.fit(df[['Area']],df.Price)
```

```python
reg.predict([[4400]])
```

```python
reg.predict([[4000]])
```

```python
reg.coef_ #value of m
```

```python
reg.intercept_ # value of b
```

y=mx+b , x is area and y is price  m is slope and b is y intercept
