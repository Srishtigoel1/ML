```python
import pandas as pd
from sklearn import linear_model
```

```python
df=pd.read_csv("homeprices.csv")
df
```

```python
medianofbedrooms=df.bedrooms.median()
medianofbedrooms
```

```python
df.bedrooms.fillna(medianofbedrooms)
```

```python
df.bedrooms=df.bedrooms.fillna(medianofbedrooms)
```

```python
df
```

```python
reg=linear_model.LinearRegression()
reg.fit(df[['area','bedrooms','age']],df[['price']])
```

```python
reg.predict([[3250,4.0,34]])
```
